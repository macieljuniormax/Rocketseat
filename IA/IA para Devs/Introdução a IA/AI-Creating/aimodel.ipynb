```python
import pandas as pd

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

### Dataset de Casas da Calif√≥rnia

```python
housing = datasets.fetch_california_housing()
housing
```

```python
dataframe = pd.DataFrame(data=housing.data, columns=housing.feature_names)
dataframe.head()
```

```python
dataframeTarget = pd.DataFrame(data=housing.target, columns=housing.target_names)
dataframeTarget.head()
```

```python
X = dataframe
Y = dataframeTarget
```

### Preparando os Dados

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=123)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Treinando o Modelo

```python
regressor = LinearRegression().fit(X_train_scaled, Y_train)
result = regressor.predict(X_test_scaled)
```

```python
result
```

```python
regressor.score(X_test_scaled, Y_test)
```
